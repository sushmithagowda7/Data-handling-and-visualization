```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data= { "Data":pd.to_datetime(["2023-05-06","2006-09-05","2016-05-06","2005-08-01","2023-05-06","2007-09-05","2023-05-06","2022-08-01",]),
       "Value":[10,16,30,37,50,65,70,78]}
```

```python
df=pd.DataFrame(data)
df.set_index("Data",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["Value"],marker='X')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series with Trend")
plt.grid(True)
plt.show()
```
