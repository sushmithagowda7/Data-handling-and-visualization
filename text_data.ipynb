```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter

text_data="She shares my dreams well i found a girst who is stronger than anyone and she is beautiful"
wordcloud=WordCloud(width=800,height=400,background_color='pink').generate(text_data)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud)
plt.title('WordCloud')
plt.show()
```

```python
word=text_data.split()
wordcounts= Counter(word)
plt.figure(figsize=(10,5))
plt.bar(wordcounts.keys(),wordcounts.values()) # Corrected line
plt.title('Histogram of Word Frequencies')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```
